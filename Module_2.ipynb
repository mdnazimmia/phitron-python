## Break and Continue Statements

```python
sum = 0

for i in range(1,11):
    sum+= i
    print(i,'is added to the sum')

print(sum)
```

```python
sum = 0

for i in range(1,11):
    print(i, 'is processing')

    if i % 2 == 0: ##even
        continue

    sum += i
    print(i,'is added to the sum')
```

```python
# Continue : goes to the top of the closest enclosing loop
mystring  = 'Sammy'
for letter in mystring:
    
    if letter == 'a':
        continue
    print(letter)
```

```python
## Break: breaks out of the current closest enclosing loop
mystring = 'Hobby'

for letter in mystring:
    if letter == 'b':
        break
    print(letter)
```

## Problem - 2 , Even, Odd, Positive and Negative

```python
n= int(input())
#inp = input()

numbers = input().split()

positive = 0; negative = 0; odd = 0; even = 0

for i in range(n):
    x = int((numbers[i])) ## i এর value প্রথমে x এ input হবে 
                            ## এবং Integer    হিসেবে ।     

  

    ## positive or negative
    if x > 0 :
        positive+=1
    elif x < 0 :
        negative+=1

    ## even or odd
    if x % 2 !=0: ## x % 2 == 1:
        odd+=1
    else:
        even+=1

print('Even:',even)
print('Odd:',odd)
print('Positive:',positive)
print('Negative:',negative)
```

## Problem - 3 Digit

```python
t = int(input())

for i in range(t):
    number = int(input())
    if number == 0:
        print(0)
        continue
    while number > 0:
        print(number%10, end=' ')
        number//=10
    print()
```
