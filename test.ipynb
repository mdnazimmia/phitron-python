## Operators in Python

```python
x = 10 ; y = 3

sum = x + y #summation
sub = x - y #subtraction
mul = x * y # multiplication 
div = round(x / y , 2) # division

rem = x % y # reminder

print(sum, sub , mul , div , rem )
```

## Comparison

## Problem : Digi Summation

```python
inp = input()
numbers = inp.split()



x = int(numbers[0])
y = int(numbers[1])
z = int(numbers[2])

last_digit_of_x = x % 10
last_digit_of_y = y % 10
last_digit_of_z = z % 10
sum = last_digit_of_x + last_digit_of_y + last_digit_of_z

print(sum)
```

```python
inp = input()
numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])

last_digit_of_x = x % 10
last_digit_of_y = y % 10
mul = last_digit_of_x * last_digit_of_y

print(mul)
```
