## Function def


```python
def sum_function(a,b):
    sum = a + b
    print(sum)

sum_function(20,20)
sum_function(12,12)

def my_function():
    print("From first input")
    print("Second one is also there")

my_function()
```

```python
def greetings():
    print('Good morning')

    return  1
result = greetings()
print(result)
```

```python
def sum(a,b):
    print(a+b)

sum(2,3)

def sum_with_return(a,b):
    return (a+b)

result=sum_with_return(2,3)
result += 4
print(sum_with_return(2,3)*3)
```
